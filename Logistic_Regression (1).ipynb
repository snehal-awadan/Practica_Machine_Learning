```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, r2_score, log_loss
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
```

```python
glass = pd.read_csv("Glass.csv")
glass
```

```python
# define X and y

# create obj of label encoder
le = LabelEncoder()

y = le.fit_transform(glass["Type"])
X = glass.drop(["Type"],axis = 1)
```

```python
# apply train_test 

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 24, test_size = 0.3, stratify = y
```

```python
# create obj for logistic

lr = LogisticRegression()
```

```python
# fit the model on lr

lr.fit(X_train, y_train)
print("coefficient are: ", lr.coef_)
print("intercept are: ", lr.intercept_)
```

```python
# predict the model


# find accuracy;
y_pred = lr.predict(X_test)
print("the accuracy is: ",accuracy_score(y_test, y_pred))

# find roc
y_pred_prob = lr.predict_proba(X_test)
print("the roc_auc_score is: ", roc_auc_score(y_test, y_pred_prob, multi_class = "ovr"))

# find log_loss;
print("the log loss is: ", log_loss(y_test, y_pred_prob))
```
